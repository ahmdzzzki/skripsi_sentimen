```python
import numpy as np
import matplotlib.pyplot as plt

# Data confusion matrix
cm = np.array([
    [47, 10,  2],
    [ 5, 66,  1],
    [ 1,  5,  7]
])

classes = ['neg', 'neu', 'pos']

plt.figure(figsize=(5, 4))
im = plt.imshow(cm, interpolation='nearest', cmap='Blues')

# Judul besar
plt.title('Confusion Matrix - Skenario P1', pad=20)

# Colorbar
plt.colorbar(im)

# Label axis
plt.xticks(np.arange(len(classes)), classes)
plt.yticks(np.arange(len(classes)), classes)

# Pindahkan xlabel ke atas
plt.xlabel('Prediction')
plt.gca().xaxis.set_label_position('top')  # pindah posisi
plt.gca().xaxis.tick_top()  # pindah tick ke atas

# Label Y
plt.ylabel('Actual')

# Tampilkan angka dalam kotak
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, str(cm[i, j]),
                 ha='center', va='center',
                 color='white' if cm[i, j] > cm.max()/2 else 'black')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data hasil evaluasi per skenario
skenario = ["P1", "P2", "P3", "P4"]

akurasi = [0.8333, 0.8125, 0.8194, 0.8264]
macro_f1 = [0.7702, 0.7100, 0.7611, 0.7328]

x = np.arange(len(skenario))
width = 0.35

plt.figure(figsize=(10, 6))

# Bar akurasi
plt.bar(x - width/2, akurasi, width, label="Akurasi")

# Bar macro F1
plt.bar(x + width/2, macro_f1, width, label="Macro F1 Score")

# Label dan judul
plt.xlabel("Skenario Preprocessing")
plt.ylabel("Nilai Evaluasi")
plt.title("Perbandingan Akurasi dan Macro F1 Score Antar Skenario")
plt.xticks(x, skenario)
plt.ylim(0, 1.0)

plt.legend()
plt.grid(axis='y', linestyle="--", alpha=0.5)

# Simpan gambar
plt.savefig("perbandingan_skenario.png", dpi=300, bbox_inches="tight")
plt.show()
```
